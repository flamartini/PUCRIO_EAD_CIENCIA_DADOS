## **MVP - Dataset para estimativa de níveis de obesidade baseada em hábitos alimentares e condição física de indivíduos da Colômbia, Peru e México.**

### Descrição do Problema:

Essa base de dados apresenta dados da estimativa de níveis de obesidade em indivíduos dos países do México, Peru e Colômbia, baseada nos seus hábitos alimentares e condição física. O Dataset contém 17 atributos e 2111 registros. Os registros são nomeados com a classe variável de resposta NObesity (Nível de obesidade), que permite classificar os dados usando valores de: Peso insuficiente, Peso Normal, Sobrepeso nível I, Sobrepeso nível II, Obesidade tipo I, Obesidade tipo II e Obesidade tipo III.
Nesta base de dados, cada linha representa uma observação de um indivíduo com relação às suas características físicas e hábitos alimentares.
Conforme normativa mexicana, de acordo com a fórmula do IMC (índice de massa corpórea), temos as seguintes classificações com relação aos níveis de obesidade:

**Mass body index** = Weight/height*height

*   **Peso insuficiente:** menos que 18.5 IMC
*   **Peso normal:** 18.5 até 24.9 IMC
*   **Sobrepeso:** 25.0 até 29.9 IMC
*   **Obesidade I:** 30.0 até 34.9 IMC
*   **Obesidade II:** 35.0 até 39.9 IMC
*   **Obesidade III:** maior que 40 IMC

Para mais detalhes sobre este dataset, consulte: (https://doi.org/10.1016/j.dib.2019.104344).
Para discussão e mais informação sobre a criação do dataset, referenciar o artigo completo “Obesity Level Estimation Software based on Decision Trees” (De-La-Hoz-Correa et al., 2019).

#### Atributos do Dataset:

Dicionário de dados: 17 atributos.

> **Atributos relacionados a hábitos alimentares:**

**FAVC**  - Consumo de alimentos altamente calóricos (Yes, No).

**SCC**    - Monitoramento de consumo de calorias (Yes, No).

**TUE**    - Tempo de uso de dispositivos tecnológicos (0–2 hours, 3–5 hours, More than 5 hours).

**MTRANS** - Transporte utilizado (Automobile, Bike, Motorbike, Public_Transportation, Walking).

> **Outras variáveis:**

**family_history_with_overweight** - histórico familiar de sobrepeso.

**Gender** - Gênero.

**Age** - Idade (em anos).

**Height** - Altura (em metros).

**Weight** - Peso (em Kg).

> **Variável de classe:**

**NObesity** - Nível de obesidade.

*   **Insufficient_Weight** - Peso insuficiente
*   **Normal_Weight**       - Peso normal

*   **Obesity_Type_II**     - Obesidade tipo II
*   **Obesidade tipo III**  - Obesidade tipo III

### Pré-processamento de dados

Alguns problemas que devem ser analisados na etapa de pré-processamento dos dados são: unidades de medida, valores faltantes, valores inconsistentes e intervalo dos valores (dimensionalidade).
Para podermos avaliar melhor o resumo estatístico do DataSet com maior precisão necessita-se transformar os dados categóricos da tabela em numéricos, e com isso podermos avaliar a média, moda, mediana e coeficiente de variação. Uma maneira popular de fazer isso se chama **codificação one-hot (OHE, one-hot encoding)**.
A OHE é uma maneira de transformarmos uma característica categórica, que pode ser composta de rótulos de texto nos dados brutos, em uma característica numérica que possa ser usada em modelos matemáticos.


Para a coluna CAEC criaremos uma coluna vazia CAEC_CAT para referenciarmos as categorias rotuladas.

```python
data_obesity['CAEC_CAT'] = 'none'
```

Analisando as primeiras linhas do DataSet referentes às colunas CAEC e CAEC_CAT:

```python
data_obesity[['CAEC', 'CAEC_CAT']].head(10)
```

Criaremos um dicionário que descreva o mapeamento das categorias de CAEC (consumo de alimentos entre as refeições) usando o código a seguir:

```python
CAEC_CAT_mapping = {"no" : 1, "Sometimes" : 2, "Frequently" : 3, "Always" : 4}

data_obesity['CAEC_CAT'] = data_obesity['CAEC'].map(CAEC_CAT_mapping)
data_obesity[['CAEC', 'CAEC_CAT']].head(10)
```

```python
CAT_mapping_yes_no = {"Yes" : 1, "yes" : 1, "No" : 0, "no" : 0}
```

Aplicando o mesmo procedimento para as demais colunas categóricas: family_history_with_overweight, FAVC, SMOKE, SCC, CALC, MTRANS.

```python
data_obesity['family_history_with_overweight_CAT'] = 'none'
data_obesity['family_history_with_overweight_CAT'] = data_obesity['family_history_with_overweight'].map(CAT_mapping_yes_no)
data_obesity[['family_history_with_overweight', 'family_history_with_overweight_CAT']].head(10)
```

```python
data_obesity['FAVC_CAT'] = 'none'
data_obesity['FAVC_CAT'] = data_obesity['FAVC'].map(CAT_mapping_yes_no)
data_obesity[['FAVC', 'FAVC_CAT']].head(10)
```

```python
data_obesity['SMOKE_CAT'] = 'none'
data_obesity['SMOKE_CAT'] = data_obesity['SMOKE'].map(CAT_mapping_yes_no)
data_obesity[['SMOKE', 'SMOKE_CAT']].head(10)
```

```python
data_obesity['SCC_CAT'] = 'none'
data_obesity['SCC_CAT'] = data_obesity['SCC'].map(CAT_mapping_yes_no)
data_obesity[['SCC', 'SCC_CAT']].head(10)
```

```python
CALC_CAT_mapping = {"no" : 0, "Sometimes": 1, "Frequently": 2}
data_obesity['CALC_CAT'] = 'none'
data_obesity['CALC_CAT'] = data_obesity['CALC'].map(CALC_CAT_mapping)
data_obesity[['CALC', 'CALC_CAT']].head(10)
```

```python
MTRANS_CAT_mapping = {"Walking" : 0, "Public_Transportation": 1, "Bike": 2, "Motorbike" : 3, "Automobile" : 4}
data_obesity['MTRANS_CAT'] = 'none'
data_obesity['MTRANS_CAT'] = data_obesity['MTRANS'].map(MTRANS_CAT_mapping)
data_obesity[['MTRANS', 'MTRANS_CAT']].head(10)
```

Agora estamos prontos para a codificação one-hot. Podemos executá-la passando uma série (Series) de um DataFrame para a função get_dummies() do pandas. O resultado será um novo DataFrame, com um número igual de colunas e níveis da variável categórica.

```python
MTRANS_ohe = pd.get_dummies(data_obesity['MTRANS_CAT'])
MTRANS_ohe.head(10)
```

**Padronização**

É o redimensionamento dos recursos para criar um “padrão” garantindo que os novos dados tenham média zero e desvio padrão igual a 1.



```python
from sklearn.preprocessing import MinMaxScaler # normalização
from sklearn.preprocessing import StandardScaler # padronização
from sklearn.preprocessing import OrdinalEncoder # ordinal encoding
from sklearn.preprocessing import OneHotEncoder # one-hot encoding e dummy encoding

scaler = StandardScaler()

# Fazendo o fit com os dados
scaler = scaler.fit(data_obesity[['Age']])

# Fazendo o transform dos dados e criando a coluna Age_padrao
data_obesity['Age_padrao'] = scaler.transform(data_obesity[['Age']])

# Agora visualizando estatisticamente esses dados
data_obesity[['Age','Age_padrao']].describe()
```

```python
scaler = StandardScaler()

# Fazendo o fit com os dados
scaler = scaler.fit(data_obesity[['Weight']])

# Fazendo o transform dos dados e criando a coluna Age_padrao
data_obesity['Weight_padrao'] = scaler.transform(data_obesity[['Weight']])

# Agora visualizando estatisticamente esses dados
data_obesity[['Weight','Weight_padrao']].describe()
```

```python
scaler = StandardScaler()

# Fazendo o fit com os dados
scaler = scaler.fit(data_obesity[['FAF']]) # NCP - número de refeições principais

# Fazendo o transform dos dados e criando a coluna Age_padrao
data_obesity['FAF_padrao'] = scaler.transform(data_obesity[['NCP']])

# Agora visualizando estatisticamente esses dados
data_obesity[['FAF','FAF_padrao']].describe()
```

```python
# Podemos inclusive verificar apenas a média
print(data_obesity.Age_padrao.mean())

# E o desvio padrão
print(data_obesity.Age_padrao.std())
```
