### **Descrição**

Embora, normalmente, Big Data esteja associada a grandes volumes de dados, sua definição formal é dada por um conjunto de três a cinco "Vs". Inicialmente a definição para "Vs" é de dados produzidos com volume, velocidade e variedade. Para dois "Vs" a mais, aparecem outras definições: veracidade e valor. O patamar de Big Data passa a exigir formas mais evoluídas de análise de dados, que agora estenderam a sua variedade, ampliaram os seus volumes e são produzidos em altíssima velocidade, é o fenômeno em que dados são produzidos em vários formatos e armazenados por uma grande quantidade de dispositivos e equipamentos.

O desafio passou a ser, por um lado, construir ferramentas e algoritmos que – sob um novo paradigma de processamento de dados – sejam capazes de lidar simultaneamente com os três "Vs" durante a análise dos dados e, por outro lado, continuamente identificar perguntas relevantes sobre o negócio, cujas respostas direcionem o pipeline de dados (coleta, integração, armazenamento, cruzamento, processamento) para subsidiar os gestores em suas análises e tomadas de decisão. Tais desafios são os pilares para uma estratégia de gestão de desempenho organizacional.

A gestão de desempenho organizacional é o conjunto de processos gerenciais e analíticos, suportados por tecnologia, que permitem ao negócio definir objetivos estratégicos e então medir e monitorar o desempenho da organização em relação a tais objetivos, através de indicadores de desempenho. Os processos analíticos da gestão de desempenho organizacional, quando direcionados por dados (data analytics), podem se dar em diversos níveis, a partir de um repositório de dados historicamente coletados sobre o negócio.

O ambiente tecnológico no qual a gestão de desempenho organizacional e seus processos analíticos são executados é denominado “ambiente de inteligência de negócio” (business intelligence). Nesses ambientes, a análise de dados nos níveis descritivo e diagnóstico é tradicionalmente suportada, envolvendo componentes como: ferramentas de apoio ao processo de extração, transformação e carga de dados a partir das fontes operacionais (ETL); o repositório integrado de dados históricos (DW, data marts e data lakes, que serão definidos mais adiante); e ferramentas de apoio ao processamento analítico on-line (OLAP).

#### **Busca pelos dados**

Dataset para estimativa de níveis de obesidade baseada em hábitos alimentares e condição física de indivíduos da Colômbia, Peru e México:

Essa base de dados apresenta dados da estimativa de níveis de obesidade em indivíduos dos países do México, Peru e Colômbia, baseada nos seus hábitos alimentares e condição física. O Dataset contém 17 atributos e 2111 registros. Os registros são nomeados com a classe variável de resposta NObesity (Nível de obesidade), que permite classificar os dados usando valores de: Peso insuficiente, Peso Normal, Sobrepeso nível I, Sobrepeso nível II, Obesidade tipo I, Obesidade tipo II e Obesidade tipo III. Nesta base de dados, cada linha representa uma observação de um indivíduo com relação às suas características físicas e hábitos alimentares. Conforme normativa mexicana, de acordo com a fórmula do IMC (índice de massa corpórea), temos as seguintes classificações com relação aos níveis de obesidade:

Mass body index = Weight/height*height

* **Peso insuficiente:** menos que 18.5 IMC
* **Peso normal:** 18.5 até 24.9 IMC
* **Sobrepeso:** 25.0 até 29.9 IMC
* **Obesidade I:** 30.0 até 34.9 IMC
* **Obesidade II:** 35.0 até 39.9 IMC
* **Obesidade III:** maior que 40 IMC

Para mais detalhes sobre este dataset, consulte: (https://doi.org/10.1016/j.dib.2019.104344). Para discussão e mais informação sobre a criação do dataset, referenciar o artigo completo “Obesity Level Estimation Software based on Decision Trees” (De-La-Hoz-Correa et al., 2019).

#### **Modelagem**

Embora o sistema relacional tenha se mostrado eficiente para suportar operações e manter dados de negócio operacionais e íntegros, o modelo não se mostrou eficiente no processo de geração de informação e conhecimento.

Na década de 1990 se popularizou então a construção de data warehouses, verdadeiros depósitos de dados, estruturados a partir de bancos de dados de operação, normalemnte relacionais. Sua estrutura difere no sentido de os dados serem armazenados no intuito de facilitar a análise, com informações pré-calculadas e dados não normalizados.

Ao contrário de um banco de dados operacional em que são feitas operações de inclusão, alteração e atualização, no data warehouse apenas se inclui dados. A redundância e a manutenção do histórico podem transformar data warehouses em gigantescos depósitos de dados.

Diferente do modelo relacional, um data warehouse utiliza o modelo multidimensional que tem como elementyo central um fato. O fato é a informação central, o tema ao qual se quer analisar. Um fato possui medidas que são valores a serem analisados e pré-calculados. Um fato também possui dimensões que são os diversos pontos de vista sobre o qual se quer analisar o fato.



**Catálogo de dados**


catalogo_dados.png


O fato no contexto das perguntas a serem respondidas nesse dataset seria o nível de obesidade, a informação central, que pode ser avaliada segundo os pontos de vista dos hábitos alimentares, de saúde e segundo as características físicas de cada indivíduo.

image.png

Para projetar um DW seguindo um **esquema estrela**, para cada fato deve haver apenas uma tabela (sem redundância alguma), e essa tabela fato deve estar relacionada a exatamente uma tabela para cada dimensão que descreve aquele fato. Isso determina que as dimensões sejam necessariamente desnormalizadas, ou seja, as hierarquias de cada dimensão são representadas por atributos na tabela dimensão.

Benefícios dos esquemas em estrela:

* Modelos de fato/dimensionais, como esquemas em estrela, são simples de entender e implementar, tornando mais fácil para que os usuários finais encontrem os dados de que precisam. Também podem ser aplicados a data marts e outros recursos de dados.
* Em comparação com modelos normalizados, como esquemas em floco de neve, os esquemas em estrela dependem menos de joins para acessar dados, tornando-os ideais para queries simples.
* Adaptam-se bem aos modelos OLAP.
* Os esquemas em estrela tentam evitar joins computacionalmente caras, resultando em melhor desempenho de query em comparação com dados normalizados.

Após a extração dos dados do dataset de níveis de obesidade, verificou-se algumas inconsistências nos tipos de dados gerados para algumas colunas e também inconsistências semânticas, conforme a definição dos metadados de negócio.

Inconsistências identificadas:

Colunas:

* **Age:** deve ser truncada pois não nos interessa idade com casas decimais;
* **Height e Weight:** devem ser decimais e com 2 casas;
* **FCVC:** primeiro devemos truncar os valores para número inteiro para depois aplicar a transformação (1 - Never; 2 - Sometimes; 3 - Always).
* **NCP, FAF:** devem ser truncados, sem casas decimais, para ficar coerente com seus metadados;
* **TUE, CH2O:** devem ser truncados, com 2 casas decimais.

Observa-se no contexto desse dataset, que como não temos uma coluna de identficação da pessoa entrevistada, como um documento ou nome ou até mesmo um identificador único, não devemos remover possíveis registros duplicados, pois pessoas diferentes podem ter registrado as mesmas características físicas e hábitos alimentares e saúde.

Segue a sequência de etapas de transformação da fase Recipe:

1. derive type: single value: trunc(FCVC, 0) as: 'FCVC_trunc'
2. settype col: FCVC_trunc lockDataType: true type: String
3. replacepatterns col: FCVC_trunc with: 'Never' on: '1'
4. replacepatterns col: FCVC_trunc with: 'Sometimes' on: '2'
5. replacepatterns col: FCVC_trunc with: 'Always' on: '3'
6. drop col: FCVC action: Drop
7. rename type: manual mapping: [FCVC_trunc,'FCVC']

18. drop col: CH2O action: Drop
19. settype col: trunc_CH2O lockDataType: true type: Float
20. rename type: manual mapping: [trunc_CH2O,'CH2O']
21. drop col: FAF action: Drop

##### **Solução do problema**

Spark é um framework utilizado para processar grandes quantidades de dados, como uma plataforma de computação em cluster com paralelismo, ou seja, os dados são distribuídos entre os clusters para processá-los de maneira paralela.

O Spark foi criado para resolver um problema de degradação de desempenho no MapReduce, uma vez que a implementação de uma aplicação usando o modelo MapReduce pode exigir muitas tarefas Map e Reduce encadeadas, as quais geram muitos arquivos intermediários no disco, em tempo de execução. Com isso, o acesso aos dados se torna intensivo, aumentando demasiadamente o custo de operações complexas e degradando o desempenho da infraestrutura como um todo.

O Spark substitui estruturas em disco por estruturas em memória distribuída, eliminando o custo de I/O associado a múltiplas leituras e escritas em disco do MapReduce.

O que antes era atualizado em arquivos em disco, agora, no Spark, é atualizado diretamente, em memória distribuída pelos nós do cluster. Dessa forma, dados intermediários do processamento são acessados pelas tarefas Map e Reduce diretamente na memória dos nós do cluster, acelerando o desempenho antes degradado dos jobs MapReduce.

Além da maior utilização da memória (em vez de disco) para execução de operações de processamento, uma das características principais do Spark é a execução de aplicações interativas, como, por exemplo, uma consulta SQL.

```python
from pyspark import SparkContext, SparkConf
from pyspark.sql import SparkSession

spark = SparkSession.builder.master("local[*]").getOrCreate()

conf = SparkConf().set('spark.ui.port', '4050').setAppName("NObesity").setMaster("local[*]")
sc = SparkContext.getOrCreate(conf=conf)
```

```python
# Importando bibliotecas

from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
# Criando o DF

url = "/content/drive/My Drive/Colab Notebooks/ObesityDataSet_tratado.csv"
df = spark.read.option("inferSchema",True).option("header",True).csv(url)

df.show()
```

```python
# Verificando valores nulos

for coluna in df.columns:
    print(coluna, df.filter(col(coluna).isNull()).count())
```

###### **1. Quais níveis de obesidade são mais afetados pela genética familiar?**

**(Genética familiar de sobrepeso = yes)**

```python
# Genética familiar de sobrepeso = yes
df.filter("family_history_with_overweight == 'yes'").groupBy('NObeyesdad').count() \
.orderBy(col('count') \
         .desc()).show(truncate=False)
```

**(Genética familiar de sobrepeso = no)**

```python
df.filter("family_history_with_overweight == 'no'").groupBy('NObeyesdad').count() \
.orderBy(col('count') \
         .desc()).show(truncate=False)
```

Verifica-se que os níveis de obesidade mais afetados pela genética familiar são os níveis mais altos.

###### **2. Pessoas fumantes tem baixos níveis de obesidade?**

```python
df.filter("SMOKE == 'yes'").groupBy('NObeyesdad').count() \
.orderBy(col('count') \
         .desc()).show(truncate=False)
```

Verifica-se que pessoas fumantes tem baixa influência nos baixos níveis de obesidade, inclusive a maioria das pessoas nos diversos níveis de obesidade são não fumantes.

###### **3. Pessoas que responderam Sim para consumo de alimentos altamente calóricos e realizam de 4 a 5 dias por semana de atividade física possuem índices menores de obesidade?**

```python
# Frequência de atividade física
# qual máximo de FAF registrado no Dataset
df.select(max(col("FAF"))).show(truncate=False)
```

O máximo de atividade física realizada por semana nessa população do dataframe é 3 vezes por semana.

**(FAVC = yes - FAF entre 2 e 3 dias por semana)**

```python
df.filter(("FAVC == 'yes'")).where(col("FAF").between(2,3)).groupBy('NObeyesdad').count() \
.orderBy(col('count') \
         .desc()).show(truncate=False)
```

As pessoas que responderam sim para consumo de alimentos altamente calóricos e se exercitam entre 2 e 3 vezes na semana são em sua maioria pessoas com mais baixos níveis de obesidade.

**(FAVC = yes - FAF entre 0 e 1 dia por semana)**

```python
df.where("FAVC == 'yes'").where(col("FAF").between(0,1)).groupBy('NObeyesdad').count() \
.orderBy(col('count') \
         .desc()).show(truncate=False)
```

Verifica-se um baixo percentual de pessoas que não consomem alimentos altamente calóricos e são obesas, a maioria das pessoas obesas consome e não praticam atividades físicas.

**(FAVC = no - FAF entre 2 e 3 dias por semana)**

```python
df.filter("FAVC == 'no'").where(col("FAF").between(2,3)).groupBy('NObeyesdad').count() \
.orderBy(col('count') \
         .desc()).show(truncate=False)
```

**(FAVC = no - FAF entre 0 e 1 dia por semana)**

```python
df.filter("FAVC == 'no'").where(col("FAF").between(0,1)).groupBy('NObeyesdad').count() \
.orderBy(col('count') \
         .desc()).show(truncate=False)
```

###### **4. Pessoas que utilizam bicicleta ou andam a pé diariamente possuem níveis mais baixos de obesidade?**

**(MTRANS = "Walking" ou MTRANS = "Bike")**

```python
df.filter((df.MTRANS == "Walking") | (df.MTRANS == "Bike")).groupBy('NObeyesdad').count() \
.orderBy(col('count') \
         .desc()).show(truncate=False)
```

Sim, podemos responder que pessoas que andam de bicicleta ou à pé diariamente possuem níveis baixos de obesidade.

###### **5. Qual é a influência do uso frequente de aparelhos eletronicos nos altos níveis de obesidade?**

```python
# máximo de horas de uso de aparelhos eletrônicos
df.select(max(col("TUE"))).show(truncate=False)
```

**(Tempo de uso de dispositivos tecnológicos)**

```python
df.select("TUE", "NObeyesdad").groupBy("TUE", "NObeyesdad").count() \
.orderBy(col('TUE') \
         .desc()).show(50, truncate=False)
```

```python
df.select("TUE", "NObeyesdad").where(col("TUE").between(0,1)).groupBy('TUE', 'NObeyesdad').count() \
.orderBy(col('count') \
         .desc()).show(50, truncate=False)
```

**(TUE entre 2 e 3 horas)**

```python
df.select("TUE", "NObeyesdad").where(col("TUE").between(2,3)).groupBy('TUE', 'NObeyesdad').count() \
.orderBy(col('count') \
         .desc()).show(truncate=False)
```

Para esta população do Dataframe, o uso frequente de aparelhos eletrônicos não teve influência nos altos níveis de obesidade, pois a maioria das pessoas com altos níveis de obesidade registrou entre 0 e 1 hora de uso por dia.

Nota-se uma faixa etária bastante jovem para os níveis mais altos de obesidade (Obesity_Type_I, Obesity_Type_II, Obesity_Type_III), entre 20 e 30 anos.

###### **7. Consumir alimentos entre as principais refeições diárias tem alto impacto nos níveis de obesidade?**

###### **8. Pessoas que responderam Sempre para "geralmente comem vegetais nas principais refeições" tem mais baixos níveis de obesidade?**

**(FCVC = "Always")**

A maioria das pessoas com o mais alto nível de obesidade (Obesity_Type_III) respondeu que sempre consome vegetais nas principais refeições, então não é um parâmetro que está ssociado à pessoas com níveis mais baixos de obesidade.

###### **9. Quais as principais caracteristicas de habitos alimentares das pessoas com níveis de obesidade Obesity II e III?**

Atributos relacionados a hábitos alimentares:

* **FAVC** - Consumo de alimentos altamente calóricos (Yes, No).
* **FCVC** - Frequência de consumo de vegetais (Never, Sometimes, Always).
* **NCP** - Número de refeições principais.
* **CAEC** - Consumo de alimentos entre as refeições (No, Sometimes, Frequently, Always).
* **SMOKE** - Fumante (Yes, No)
* **CH2O** - Consumo diário de água (em litros).
* **CALC** - Consumo de álcool (Always, Frequently, No, Sometimes).

**Hábitos alimentares Obesity_Type_II**

```python
df.select('FAVC', 'FCVC', 'CAEC', 'SMOKE', 'CALC', 'NObeyesdad') \
.filter(df.NObeyesdad == "Obesity_Type_II").groupBy('FAVC', 'FCVC', 'CAEC', 'SMOKE', 'CALC', 'NObeyesdad') \
.count().orderBy(col('count') \
.desc()).show(50, truncate=False)
```

**(NCP entre 0 e 1 refeição principal)**

```python
# NCP - Número de refeições principais.
df.select("NCP", "NObeyesdad").where(col("NCP").between(0,1)).groupBy('NCP', 'NObeyesdad').count() \
.orderBy(col('count') \
         .desc()).show(truncate=False)
```

**(NCP entre 2 e 3 refeições principais)**

```python
# NCP - Número de refeições principais.
df.select("NCP", "NObeyesdad").where(col("NCP").between(2,3)).groupBy('NCP', 'NObeyesdad').count() \
.orderBy(col('count') \
         .desc()).show(truncate=False)
```

**(NCP entre 4 e 5 refeições principais)**

```python
# NCP - Número de refeições principais.
df.select("NCP", "NObeyesdad").where(col("NCP").between(4,5)).groupBy('NCP', 'NObeyesdad').count() \
.orderBy(col('count') \
         .desc()).show(truncate=False)
```

**(CH2O entre 0 e 1 litro diário)**

```python
df.select("CH2O", "NObeyesdad").where(col("CH2O").between(0,1)).groupBy('CH2O', 'NObeyesdad').count() \
.orderBy(col('count') \
         .desc()).show(50, truncate=False)
```

**(CH2O entre 2 e 3 litros diários)**

```python
df.select("CH2O", "NObeyesdad").where(col("CH2O").between(2,3)).groupBy('CH2O', 'NObeyesdad').count() \
.orderBy(col('count') \
         .desc()).show(50, truncate=False)
```

**(CH2O maior que 4 litros diários)**

```python
df.select("CH2O", "NObeyesdad").where('CH2O > 4').groupBy('CH2O', 'NObeyesdad').count() \
.orderBy(col('count') \
         .desc()).show(50, truncate=False)
```

**(Média de CH2O por nível de obesidade)**

```python
df.groupBy('NObeyesdad').agg({'CH2O' : 'avg'}).show(50, truncate=False)
```

**Hábitos alimentares Obesity_Type_III**

```python
df.select('FAVC', 'FCVC', 'CAEC', 'SMOKE', 'CALC', 'NObeyesdad') \
.filter(df.NObeyesdad == "Obesity_Type_III").groupBy('FAVC', 'FCVC', 'CAEC', 'SMOKE', 'CALC', 'NObeyesdad') \
.count().orderBy(col('count') \
.desc()).show(50, truncate=False)
```

A maioria das pessoas dos 2 níveis de obesidade mais altos respondeu:

image.png

Com isso não é possível verificar um padrão característico de hábitos alimentares ruins para as pessoas com os mais altos níveis de obesidade, com exceção do atributo **FAVC** que mostra o hábito de consumo de alimentos altamente calóricos.

Em média o consumo diário de água em todos os níveis de obesidade foi 2 litros e o número total de refeições diárias foi 3, o que são bons parâmetros para níveis de saúde.

###### **10. Quais as principais caracteristicas de hábitos alimentares das pessoas com níveis de obesidade Underweight e Normal?**

**(Hábitos alimentares NObeyesdad = "Insufficient_Weight")**

```python
df.select('FAVC', 'FCVC', 'CAEC', 'SMOKE', 'CALC') \
.filter(df.NObeyesdad == "Insufficient_Weight").groupBy('FAVC', 'FCVC', 'CAEC', 'SMOKE', 'CALC') \
.count().orderBy(col('count') \
.desc()).show(50, truncate=False)
```

**(Hábitos alimentares NObeyesdad = "Normal_Weight")**

```python
df.select('FAVC', 'FCVC', 'CAEC', 'SMOKE', 'CALC') \
.filter(df.NObeyesdad == "Normal_Weight").groupBy('FAVC', 'FCVC', 'CAEC', 'SMOKE', 'CALC') \
.count().orderBy('FAVC', col('count') \
.desc()).show(50, truncate=False)
```

Conclui-se que as pessoas abaixo do peso ideal tem maior necessidade de consumo de alimentos alltamente calóricos, enquanto as pessoas que estão com seu peso normal/adequado, não consomem alimentos altamente calóricos em sua maioria.
